```python
def full_convert(s):
    s = str(s)
    s=s.lower()
    if s=='l' or s=='i':
        return 18
    if 'pm' in s:
        s = s[:s.index('pm')]
    if '&' in s:
        s = s.replace("&",'')
        s=min(s)
    if 'oclk' in s:
        s = s[:s.index('oclk')]
    if s.isdigit():
        if int(s)<9:
            return int(s)+12
        else:
            return s
    return s
    
    
        
        
```

```python
def type_of_leave(n):
    if n<=-2:
        return "Informed Leave"
    else:
        return "Emergency Leave"
data2['leave type'] = data2['Cancellation before'].apply(type_of_leave)
```
